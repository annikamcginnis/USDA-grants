```python
import pandas as pd 

datasets = ['usda_assistance_2025_2022.csv', 'df_2021_2020.csv', 'usda_assistance_2019_2015.csv']
output_file = 'usda_assistance_2015_2025.csv'

# Open output file for writing
with open(output_file, 'w', encoding='utf-8') as f_out:
    first_file = True  # To handle writing the header only once

    for file in datasets:
        chunk_iter = pd.read_csv(file, chunksize=100000)  # Read in chunks of 100,000 rows

        for chunk in chunk_iter:
            chunk.to_csv(f_out, index=False, header=first_file)  # Write to file, header only for first file
            first_file = False  # After first file, don't write header

print("Merging complete. Data saved to:", output_file)
```
